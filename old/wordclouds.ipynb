```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from gensim.corpora import Dictionary
from gensim.models import LdaMulticore

lda_model = LdaMulticore.load('start-2021-06-24/trained_lda')
```

```python
# generate word cloud
for t in list(range(lda_model.num_topics))[5:10]:
    plt.figure()
    plt.imshow(WordCloud(background_color='white').fit_words(dict(lda_model.show_topic(t, 200))))
    plt.axis("off")
    plt.title("Topic #" + str(t))
    # plt.savefig(save_dirname+'/topic'+str(t)+'wordcloud.pdf', format='pdf')
    # plt.close()
    # break
```
